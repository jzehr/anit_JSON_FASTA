```r
install.packages("ggplot2")
library(ggplot2)
```

```r
d <- read.csv(file = 'TSL7,9,11.tsv', sep = '\t')
```

```r
d
```

```r
ggplot(d, aes(CDR3.Length..AA., fill = Time, color = Time))+
  geom_density(alpha = .3)
```

```r
ggplot(data = d) +
    geom_point(mapping = aes(Time,Size, color = Time)) +
    facet_wrap(~v_region)
ggsave("IgG_V3_full:7,9,11.png", width = 13, height =8)
```

```r
ggplot(data = d) +
    geom_point(mapping = aes(Time,Size, color = Time)) +
    facet_wrap(~v_gene)
ggsave("IgG_V3_gene:7,9,11.png", width = 13, height =8)
```
